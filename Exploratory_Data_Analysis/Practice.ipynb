#### 5.   `most_freq_aspect` 별로  `reviews_per_month`의 차이가 존재하는가? Room type 별로 비교하여라. (그룹별 평균을 barplot으로 비교)

```python
fig, axes = plt.subplots(figsize = (12,7))
sns.barplot(x='most_freq_aspect', y='reviews_per_month', hue = 'room_type', data = data, estimator=np.mean ,ax = axes, ci=95, capsize=0.2)
plt.show()
```

```python
sorted_list = sorted(data.query("room_type == 'Entire home/apt'")['most_freq_aspect'].unique())

data.most_freq_aspect= data.most_freq_aspect.astype('category')
data.most_freq_aspect = data.most_freq_aspect.cat.reorder_categories(sorted_list)
```

```python
# Room_Type = Entire home/apt

data_entire = data.query("room_type == 'Entire home/apt'")

fig, axes = plt.subplots(1, 2, figsize = (30,12))
sns.barplot(x='most_freq_aspect', y='reviews_per_month', data = data_entire, estimator=np.mean ,ax = axes[0], ci=95, capsize=0.2)
sns.violinplot(x = 'most_freq_aspect', y ='reviews_per_month', data = data_entire, ax = axes[1])

axes[0].tick_params(axis = 'x', labelsize= 20, rotation = 45)
axes[0].tick_params(axis = 'y', labelsize= 20)
axes[0].set_xlabel('Most_freq_aspect', fontsize = 25)
axes[0].set_ylabel('Reviews_per_Month', fontsize = 25)

axes[1].tick_params(axis = 'x', labelsize= 20, rotation = 45)
axes[1].tick_params(axis = 'y', labelsize= 20)
axes[1].set_xlabel('Most_freq_aspect', fontsize = 25)
axes[1].set_ylabel('Reviews_per_Month', fontsize = 25)

plt.suptitle('Room Type - Entire Home/apt.',fontsize=30, fontweight = 'bold', y = 0.99)
plt.show()
```

```python
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm


model = smf.ols('reviews_per_month ~ most_freq_aspect', data = data_entire).fit()
display(model.summary().tables[1])
display(anova_lm(model))
```

#### 2. 평활상수 0.5를 사용하여 지수가중이동평균(EWMA)을 사용하여 대여횟수를 평활하여라.  2012년 12월 25일의 예측오차는 무엇인가?

#### 3. 대여횟수를 시계열분해하여 오차를 변수명 'resid'로 저장하시오. 

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(bike.cnt, model='additive', 
                            period=7)
```

```python
result.plot()
plt.show()
```

```python
bike['resid'] = result.resid
```

#### 4.`resid` 변수의 ACF, PACF 그림을 그리고 ARMA 모형을 사용한다면 그 차수를 판별하시오. 

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig, axes = plt.subplots(1,2, figsize=(12,3))
plot_acf(bike.resid[3:-3],ax=axes[0])
plot_pacf(bike.resid[3:-3], ax=axes[1] )
plt.show()
```

- ACF 가 lag 3 이후에 Cut-off
- PACF 가 Tail off q = 3 이기 때문에 MA(3) 모형

```python
# 시차가 1이므로, 이전 행과 시차를 계산
data['Time Stamp'].diff()
```

```python
def add_newsession(df):
    df['Newsession'] = df['Time Stamp'].diff() > pd.Timedelta(minutes=20) # 서로다른 아이디가 앞뒤로 있으면 
    df['Newsession'].iloc[0] = True # 같은 Session으로 count 되어서 이렇게 바꿔야 한다 = ID의 첫 세션을 True로 바꿔주는 것
    return df
    
data = data.groupby('KMID').apply(add_newsession)    
```
